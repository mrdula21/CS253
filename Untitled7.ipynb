```python
#Basically, we just convert the party names to integers
from sklearn.preprocessing import LabelEncoder
parties = df['Party'].tolist()

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit the LabelEncoder to the list of parties and transform them into integers
party_integers = label_encoder.fit_transform(parties)
party_name_mapping = dict(zip(label_encoder.transform(label_encoder.classes_), label_encoder.classes_))

# Update the 'party_integer' column in x_train with the encoded values
df['Party'] = party_integers
```

```python
#Converting states to integers
parties = df['state'].tolist()

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit the LabelEncoder to the list of parties and transform them into integers
party_integers = label_encoder.fit_transform(parties)

# Update the 'party_integer' column in x_train with the encoded values
df['state'] = party_integers
```

```python
#constituency to integers
parties = df['Constituency ∇'].tolist()

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit the LabelEncoder to the list of parties and transform them into integers
party_integers = label_encoder.fit_transform(parties)

# Update the 'party_integer' column in x_train with the encoded values
df['Constituency ∇'] = party_integers
```

```python
#converting string values of liabilities and assets to float
import numpy as np
def parse_value(value_str):
    if "Lac" in value_str:
        value = np.log(float(value_str.replace(" Lac+", "").strip()) * 100000)
    elif "Crore" in value_str:
        value = np.log(float(value_str.replace(" Crore+", "").strip()) * 10000000)
    else:
        value =0

    return value

df["Total Assets"] = df["Total Assets"].apply(parse_value)
df["Liabilities"] = df["Liabilities"].apply(parse_value)
```

```python
#seperating those who are doctors and advocates
def categorize_name(name):
    lower_name = name.lower()
    if lower_name.startswith('dr.'):
        return 1
    elif lower_name.startswith('adv.'):
        return 2
    else:
        return 0

# Apply the function to the 'Name' column
df['Candidate'] = df['Candidate'].apply(categorize_name)
```

```python
# EDA
import matplotlib.pyplot as plt
party_criminal_records_sum = {}
for candidate in df.iterrows():
    party = candidate[1]["Party"]
    criminal_cases = candidate[1]["Criminal Case"]
    party_criminal_records_sum[party] = party_criminal_records_sum.get(party, 0) + criminal_cases

party_names = [party_name_mapping[party] for party in party_criminal_records_sum.keys()]

# Plot the graph
#parties = list(party_criminal_records_sum.keys())
criminal_records = list(party_criminal_records_sum.values())
plt.figure(figsize=(20, 20))  # Aspect ratio 1:1 to make it a circle
plt.pie(criminal_records, labels=party_names, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.title(' Criminal Records per Candidate by Party')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators=110, criterion='entropy',class_weight='balanced')
clf.fit(x_train, y_train)
```
